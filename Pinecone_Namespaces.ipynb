Using Namespaces

```python
import os 
from dotenv import load_dotenv, find_dotenv 
from pinecone import Pinecone  
import random
```

```python
pc=Pinecone()

index=pc.Index("langchain")

print(index.describe_index_stats())
```

```python
vectors=[[random.random() for i in range(1536)] for v in range(3)] 

ids=list("fgh") 

index.upsert(vectors=zip(ids, vectors), namespace="first_namespace")
```

```python
vectors_2=[[random.random() for i in range(1536)] for v in range(2)] 

ids_2=list("ijk") 

index.upsert(vectors=zip(ids_2, vectors_2), namespace="second_namespace")
```

```python
print(index.describe_index_stats())
```

```python
index.fetch(ids=["k"])

# We have to specify the namespace that we want to fetch from, by default we fetch from the default/empty namespace
```

```python
index.fetch(ids=["h"], namespace="first_namespace")
```
